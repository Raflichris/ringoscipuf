```python
#convert respon df1 dan df2 ke txt (masih hexa)
import numpy as np
array = df1.to_numpy()
array1 = df2.to_numpy()
np.savetxt("respon1.txt", array, fmt = "%s")
np.savetxt("respon2.txt", array1, fmt = "%s")
```

```python
#convert ke biner dari hexa ke biner format txt df1
with open("converted1.csv", 'w') as out_file, open("respon1.txt") as in_file:
    for hex in in_file.read().split():
        def hexadecimal_to_binary():
            z = int(hex, 16)
            string1 = " "
            while z > 0:
              string1 = (str(z % 2) + string1)
              z = z >> 1 #using bitwise operator to shift value by 1
            print(string1, file=out_file)
        hexadecimal_to_binary()
```

```python
#convert ke biner dari hexa ke biner format txt df 2
with open("converted2.csv", 'w') as out_file, open("respon2.txt") as in_file:
    for hex in in_file.read().split():
        def hexadecimal_to_binary():
            z = int(hex, 16)
            string1 = " "
            while z > 0:
              string1 = (str(z % 2) + string1)
              z = z >> 1 #using bitwise operator to shift value by 1
            print(string1, file=out_file)
        hexadecimal_to_binary()
```

```python
#randomness dan stadiness intra df1
import pandas as pd
import numpy as np
from scipy.spatial.distance import hamming

convdf1=pd.read_csv('converted1.csv')
arrcon1 = convdf1.to_numpy()

listbesarrandom = []
listrandom = []
liststad = []
listbesarstad=[]

i = 0
j = 1
while i < 500 :
    temp = j
    while j < 500 :
        a = hamming(list(arrcon1[i][0]), list(arrcon1[j][0])) * len(arrcon1[i][0])
        if a==0.0:
          listbesarstad.append(a)
        else:
          listrandom.append(a)
        j += 1
    listbesarrandom.append(listrandom)
    listbesarstad.append(liststad)
    listrandom = []
    liststad = []
    i += 1
    j = temp + 1
```

```python
#Print Output
print(listbesarrandom[0])
print('count randomness line 1 :', len(listbesarrandom[0]))
print(listbesarrandom[1])
print('count randomness line 2 :', len(listbesarrandom[1]))
print(listbesarrandom[2])
print('count randomness line 3', len(listbesarrandom[2]))
```

```python
def CountFrequency(my_list):
    freq = {}
    for items in my_list:
        freq[items] = my_list.count(items)
    for key, value in freq.items():
        print ("% d : % d"%(key, value))
```
